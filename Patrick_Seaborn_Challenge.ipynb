```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
sns.set_theme(style="darkgrid")
```

```python
db=pd.read_csv('Diabetes.csv')
```

```python
db.head()
```

```python
sns.scatterplot(x=db['BMI'], y=db[('Glucose')])
plt.show()
```

```python
sns.relplot(data=db, x="BMI", y="Glucose", hue="Pregnancies")  
```

```python
sns.relplot(data=db, x="BMI", y="Glucose", hue="Pregnancies", size="Pregnancies")  
```

```python
sns.relplot(data=db, x="BMI", y="Glucose", hue="Pregnancies", size="Pregnancies", sizes=(10,50))  
```

```python
sns.relplot(data=db, x="BMI", y="Glucose", hue="Pregnancies",size="Pregnancies", sizes=(10,50))  
plt.xlim(18,37)
plt.ylim(50,140)
plt.show()
```

```python
sns.relplot(data=db, x="Age", y="Insulin", hue="Pregnancies", col="Pregnancies", sizes=(10,50))
```

Via assigning the col variable to the pregnancies column, a series of scatterplot graphs are created, this can provide a less-clustered,  individual-level inspections of the data.  

```python
sd=pd.read_csv('Salary Data.csv')
```

```python
sd.head()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Salary", data=sd)
plt.title('age_salary')
plt.show()
```

```python
educ={"Bachelor's": "green", "Master's": "red","PhD": "yellow" }
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender",data=sd, palette=educ)
plt.title('age_salary@edu lvl')
plt.show()
```

```python
educ={"Bachelor's": "green", "Master's": "red","PhD": "yellow" }
gm={"Female":"o", "Male":"s"}
plt.figure()
sns.set_style("ticks")
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender",data=sd, palette=educ, markers=gm)
plt.legend(ncol=2, frameon=False, loc="lower right")
plt.title('age_salary@edu lvl')
plt.show()
```

```python
educ={"Bachelor's": "green", "Master's": "red","PhD": "yellow" }
gm={"Female":"o", "Male":"s"}
plt.figure()
sns.set_style("ticks")
ba=sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender",data=sd[sd["Education Level"]=="Bachelor's"], palette=educ, markers=gm)
ba.set_xlim(25,50)
ba.set_ylim(50000,250000)
#plt.legend(ncol=2, frameon=False, loc="lower right")
plt.title('age_salary@edu lvl')
plt.show()
```
