```python
demo = pd.read_csv('cces_example_data.csv')
```

```python
demo.head()
```

```python
fig1, ax1 = plt.subplots(figsize=(5, 2), layout='constrained')
ax1.set_xlabel('Age Histogram')
ax1.set_ylabel('Political Identity')
ax1.set_title("The Empty Figure")

plt.show()
```

```python
agm = demo.groupby('age')['Evaluation'].mean().reset_index()

fig2, ax2 = plt.subplots(figsize = (12,5))
ax2.bar(agm['age'], agm['Evaluation'], color='green')
ax2.set_xlabel('Age')
ax2.set_ylabel('Average Evaluation')
ax2.set_title('Average Evaluation by Age')

plt.show()
```

```python
fig3, ax3 = plt.subplots(figsize = (10,7))
ax3.scatter(x = demo['ImmiPhobia'], y = demo['ConMediaUse'],alpha=0.5)
ax3.set_xlabel('ConMediaUse')
ax3.set_ylabel('ImmiPhobia')
ax3.set_title('ConMediaUse vs. ImmiPhobia')
plt.grid(True)
plt.show()
```
